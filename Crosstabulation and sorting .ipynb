```python
# Pivoting 
# Create a new derivate table out of an existing one 
# Uses unique values from index/columns 
# Reshape data based on column values

data1 = pd.read_csv('C:\\Users\\USUARIO\\Desktop\\JAVO\\Python for Data Analisis\\scriptsLecture\\scriptsLecture\\section5\\GlobalFirePower.csv')
data1.head(5)
```

```python
# Specify a unique value index ISO3 is a column with all unique values
p = pd.pivot_table(data1,index=['ISO3'])
p.head(5)
```

```python
# Include the corresponding values 
p = pd.pivot_table(data1,index=['Country'],values=['Attack Aircraft'])
p.head(4)
```

```python
# 3 indices to pivot on #input 2 values 
p = pd.pivot_table(data1,index=['Country','ISO3','Rank'],values=['Attack Aircraft','Active Personnel'])
p.tail(8)
```
