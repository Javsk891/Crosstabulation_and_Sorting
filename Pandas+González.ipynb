1 - Construir un objeto de series de tiempo 

```python
import pandas as pd   # Importamos la librería de pandas con el nombre pd
import numpy as np    # Importamos la librería de numpy con el nombre np
```

```python
df = pd.read_csv('Summer-Olympic-medals-1976-to-2008.csv', encoding = 'latin-1')
```

```python
indice = list(df.Year)    # Aquí se definen los años como índeces del DataFrame
columnas = df.columns[1:]   # Se definen las columnas para el arreglo del DataFrame
valores = df.values[:,1:]   # Una vez establecidos los índices y colunas se procede a definir cuales serán los valores que relacionan los años y cada una de las columnas en DataFrame
df_olympic_medals = pd.DataFrame(valores, index=indice, columns = columnas)   # Se arma el DtaFrame con las columnas, los índices y valores 
df_olympic_medals   # Llamamos el DataFrame 'df_olympic_medals'
```

2. Separar nombres de apellidos en dos columnas distintas usando la coma como separador
